```python
# Bloco para importação das bibliotecas necessárias
import pandas as pd  # importação do pandas para manipulação de dados
from sklearn.tree import DecisionTreeClassifier, plot_tree  # para criar e visualizar a árvore de decisão.
from sklearn.metrics import accuracy_score  # para calcular a acurácia do modelo.
from sklearn.model_selection import train_test_split  # para dividir os dados em conjuntos de treinamento e teste.
import matplotlib.pyplot as plt  # para visualizar a árvore de decisão.
from sklearn import svm  # para criar e treinar o classificador SVM.

# Função para calcular e imprimir a acurácia do modelo
def imprimir_desempenho(clf, X, y, nome_modelo="Modelo"):  # recebe o classificador clf, os dados X, os rótulos y e o nome do modelo.
    previsoes = clf.predict(X)  # armazena as previsões feitas pelo classificador.
    acuracia = accuracy_score(y, previsoes)  # calcula a acurácia comparando as previsões com os rótulos verdadeiros.
    print(f"Acurácia do {nome_modelo}: {acuracia:.4f}")  # imprime a acurácia formatada com 4 casas decimais.

# Função para visualizar a árvore de decisão
def exibir_arvore_decisao(clf, nomes_caracteristicas, nomes_classes):  # recebe o classificador clf, os nomes das características e das classes.
    plt.figure(figsize=(20, 10))  # define o tamanho da figura.
    plot_tree(clf, feature_names=nomes_caracteristicas, class_names=nomes_classes, filled=True)  # visualiza a árvore de decisão.
    plt.show()  # exibe a figura.
    # Pausa para o usuário visualizar a árvore antes de continuar
    plt.pause(0.1)  # pausa para 0.1 segundos.

# Função para prever se o câncer é maligno ou benigno
def prever_cancer(clf, nomes_caracteristicas):  # recebe o classificador clf e os nomes das características.
    dados = {}  # cria um dicionário vazio para armazenar os dados do usuário.
    for caracteristica in nomes_caracteristicas:  # percorre os nomes das características.
        while True:
            # tratamento de exceção
            try:
                valor = float(input(f"Informe o valor para {caracteristica}: "))  # solicita ao usuário o valor para a característica atual.
                dados[caracteristica] = valor  # adiciona o valor ao dicionário.
                break
            except ValueError:
                print("Valor inválido. Por favor, insira um número.")
    dados_df = pd.DataFrame([dados], columns=nomes_caracteristicas)  # cria um DataFrame com os dados do usuário.
    previsao = clf.predict(dados_df)  # preve se o câncer é maligno ou benigno usando o classificador.
    if previsao[0] == 'B':
        print("A previsão para o câncer é: Benigno") # se for 'B', imprime "Benigno"
    elif previsao[0] == 'M':
        print("A previsão para o câncer é: Maligno") # se for 'M', imprime "Maligno"
    else:
        print("Erro ao prever o câncer.")

# Função Principal
def main():
    # Carregar os dados do arquivo CSV
    df = pd.read_csv("wdbc.csv")

    # Renomear as colunas para português
    df.rename(columns={
        'radius1': 'raio1', 'texture1': 'textura1', 'perimeter1': 'perímetro1', 'area1': 'área1', 'smoothness1': 'suavidade1',
        'compactness1': 'compacidade1', 'concavity1': 'concavidade1', 'concave_points1': 'pontos_côncavos1', 'symmetry1': 'simetria1',
        'fractal_dimension1': 'dimensão_fractal1', 'radius2': 'raio2', 'texture2': 'textura2', 'perimeter2': 'perímetro2', 'area2': 'área2',
        'smoothness2': 'suavidade2', 'compactness2': 'compacidade2', 'concavity2': 'concavidade2', 'concave_points2': 'pontos_côncavos2',
        'symmetry2': 'simetria2', 'fractal_dimension2': 'dimensão_fractal2', 'radius3': 'raio3', 'texture3': 'textura3', 'perimeter3': 'perímetro3',
        'area3': 'área3', 'smoothness3': 'suavidade3', 'compactness3': 'compacidade3', 'concavity3': 'concavidade3', 'concave_points3': 'pontos_côncavos3',
        'symmetry3': 'simetria3', 'fractal_dimension3': 'dimensão_fractal3'
       }, inplace=True)

    # Selecionar as colunas renomeadas
    colunas = [
        'raio1', 'textura1', 'perímetro1', 'área1', 'suavidade1',
        'compacidade1', 'concavidade1', 'pontos_côncavos1', 'simetria1',
        'dimensão_fractal1', 'raio2', 'textura2', 'perímetro2', 'área2',
        'suavidade2', 'compacidade2', 'concavidade2', 'pontos_côncavos2',
        'simetria2', 'dimensão_fractal2', 'raio3', 'textura3', 'perímetro3',
        'área3', 'suavidade3', 'compacidade3', 'concavidade3', 'pontos_côncavos3',
        'simetria3', 'dimensão_fractal3'
    ]

    # Separar características (features) e alvo (target)
    X = df[colunas]
    y = df["Diagnosis"]

    # Dividir os dados em conjuntos de treinamento e teste
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

    # Criar e treinar o classificador de árvore de decisão
    clf_arvore = DecisionTreeClassifier()  # cria um classificador de árvore de decisão.
    clf_arvore.fit(X_train, y_train)  # treina o classificador com os dados de treinamento.

    # Criar e treinar o classificador SVM
    clf_svm = svm.SVC()  # cria um classificador SVM.
    clf_svm.fit(X_train, y_train)  # treina o classificador com os dados de treinamento.

    # Estrutura de repetição para o menu
    while True:
        print("\nSeja Bem-Vindo! Esse programa foi feito para que uma equipe médica \npossa descobrir se um cancêr de mama é Maligno ou Benigno.")
        print("\nEscolha uma opção:")
        print("1 - Ver desempenho do modelo de Árvore de Decisão")
        print("2 - Ver árvore de decisão")
        print("3 - Prever câncer com Árvore de Decisão")
        print("4 - Ver desempenho do modelo SVM")
        print("5 - Prever câncer com SVM")
        print("6 - Encerrar programa")

        opcao = input("Opção: ")

        if opcao == '1':
            imprimir_desempenho(clf_arvore, X_test, y_test, nome_modelo="Árvore de Decisão")
        elif opcao == '2':
            exibir_arvore_decisao(clf_arvore, nomes_caracteristicas=colunas, nomes_classes=["Benigno", "Maligno"])
        elif opcao == '3':
            prever_cancer(clf_arvore, nomes_caracteristicas=colunas)
        elif opcao == '4':
            imprimir_desempenho(clf_svm, X_test, y_test, nome_modelo="SVM")
        elif opcao == '5':
            prever_cancer(clf_svm, nomes_caracteristicas=colunas)
        elif opcao == '6':
            print("Encerrando o programa...")
            break
        else:
            print("Opção inválida. Escolha novamente.")

if __name__ == "__main__":
    main()
```
